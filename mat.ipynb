```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(10)
y1=x**2
y2=x+3
plt.xlabel('x aixs')
plt.ylabel('y aixs')
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
x=np.arange(2)
y1=x**2
y2=2*x+3
plt.scatter(x,y2,color='green',marker='*')
plt.xlabel('x aixs')
plt.ylabel('y aixs')
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
x=np.arange(0,15)
y=x**3+3*x**+3*x+1
plt.scatter(x,y,color='red',marker='*')
plt.show()
plt.plot(x,y)
plt.show()
```

```python
n=np.arange(20)
x=np.cos(2*n)
y=np.sin(2*n)
plt.plot(n,x)
plt.plot(x,y)
plt.plot(n,y)
plt.scatter(n,y,color="red",marker="*")
plt.show()

```
